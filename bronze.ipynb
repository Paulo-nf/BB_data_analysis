```python
import pandas as pd

df = pd.read_csv('data/databridge_squad19_sintetico.csv')
```

# padronizando para uppercase

```python
df["doc_type"] = df["doc_type"].str.upper()
df["source_system"] = df["source_system"].str.upper()
df["text_language"] = df["text_language"].str.upper()
df["ocr_engine"] = df["ocr_engine"].str.upper().str.strip()
```

```python
df.to_csv("databridge_squad19_bronze.csv", index=False)
```
